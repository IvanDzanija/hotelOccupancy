# Pregled broja rezervacija po drzavi za hotel 1(id 0)

Prvo pronalazimo datoteku sa podatcima za hotel 1 i ucitavamo potrebne biblioteke.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

curr = Path.cwd()
curr = str(curr)
curr += "/country_data/no_of_reservations_per_country_hotel0.csv"
print(curr)
```

```python
df = pd.read_csv(curr)
df.info()
```

```python
df2 = df[df["Zemlja_gosta"] == "DEU"]
a = df.iloc[0].to_list()
a = a[2:]
b = df2.columns.to_list()
b = b[2:]
#print(b)
c = []
d = []
ukupno = 0
for i in range(len(a)):
    if a[i] > 0:
        c.append(a[i])
        d.append(b[i])
    ukupno += a[i]
plt.figure(figsize=(70,6))
plt.plot(d,c)
plt.xticks(rotation=60)
print(ukupno)
```
