# Obrada podataka po drzavama

U ovoj biljeznici nalaze se postupci obrade podataka uzimajuci u obzir zemlju dolaska gostiju. Pokusavamo vidjeti koje su navike i u kojem broju hotele posjecuju ljudi razlicitih zemalja.

Prvo pocinjemo uvozom potrebnih datoteka

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

Zatim moramo otvoriti datoteku gdje nam se nalaze podatci, odnosno dohvatiti njenu lokaciju

```python
curr = Path.cwd()
path = curr.parent.parent
file = str(path)
file += "/cleaned_data.csv"
print(file)
```

Sada kada nam je put to datoteke dobar mozemo ju i otvoriti

```python
df = pd.read_csv(file)
df.info()
```

Sada radimo skup svih drzava iz skupa podataka, i radimo listu dataframeova koji sadrze podatke za svaku drzavu.

```python
df['zemlja_gosta'] = df['zemlja_gosta'].replace('CN', 'CHN')
drzave = set(df["zemlja_gosta"])
d = list(drzave)
d.sort()
#print(d)
lista_df = list()
for drz in d:
    temp_df = df[df["zemlja_gosta"] == drz ]
    lista_df.append(temp_df)
    #temp_df.info()
```

Radimo numpy nizove za niz drzava i drugi za broj rezervacija iz neke drzave te radimo graf, raditi cemo vise kategorija izmedju 0-100,100-1000,1000 i vise
Prvo za 0-100

```python
np_drz = np.array([])
#print(a)
np_broj_posjetitelja = np.array([])
for DF in lista_df:
    if len(DF) < 100:
        a = list(DF["zemlja_gosta"])
        np_drz = np.append(np_drz,a[0])
        np_broj_posjetitelja = np.append(np_broj_posjetitelja,len(DF))
print(np_drz)
print(np_broj_posjetitelja)
plt.figure(figsize=(100,6))
plt.bar(np_drz,np_broj_posjetitelja)
```

Izmedju 100 i 1000

```python
np_drz = np.array([])
#print(a)
np_broj_posjetitelja = np.array([])
for DF in lista_df:
    if len(DF) > 100 and len(DF) < 1000:
        a = list(DF["zemlja_gosta"])
        np_drz = np.append(np_drz,a[0])
        np_broj_posjetitelja = np.append(np_broj_posjetitelja,len(DF))
print(np_drz)
print(np_broj_posjetitelja)
plt.figure(figsize=(20,6))
plt.bar(np_drz,np_broj_posjetitelja)
```

1000 nadalje

```python
np_drz = np.array([])
#print(a)
np_broj_posjetitelja = np.array([])
for DF in lista_df:
    if len(DF) > 1000:
        a = list(DF["zemlja_gosta"])
        np_drz = np.append(np_drz,a[0])
        np_broj_posjetitelja = np.append(np_broj_posjetitelja,len(DF))
print(np_drz)
print(np_broj_posjetitelja)
plt.figure(figsize=(20,6))
plt.bar(np_drz,np_broj_posjetitelja)
```
